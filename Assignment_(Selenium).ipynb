```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
```

```python
#opening the naukri webpage on automated Chrome browser.
driver.get("https://www.naukri.com/")
```

```python
#entering designation as required in question.
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys("Data Scientist")
```

```python
#clicking on search button after entering designation.
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
#selecting the location filter as 'Delhi/NCR' as required in question.
location=driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[3]/div[2]/div[2]/label")
location.click()
```

```python
#selecting the salary filter as '3-6 lakh' as required in question.
salary=driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[3]/div[2]/div[2]/label")
salary.click()
```

```python
#creating lists in which we are going to scrap data.
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#scraping the data of job titles.
title_tags=driver.find_elements(By.XPATH,'//a[@class="title "]')
for i in title_tags:
    job_title.append(i.text)
    
```

```python
#scraping the data of job location.
location_tags=driver.find_elements(By.XPATH,'//span[@class="locWdth"]')
for i in location_tags:
    job_location.append(i.text)
    
```

```python
#scraping the data of company's name.
company_tags=driver.find_elements(By.XPATH,'//a[@class=" comp-name mw-25"]')
for i in company_tags:
    company_name.append(i.text)
    
```

```python
#scraping the data of experience required for the job.
experience_tags=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in experience_tags:
    experience_required.append(i.text)
    
```

```python
#creating the DataFrame from above scraped data.This dataframe is having data for first 10 job result only as required in question.
df=pd.DataFrame({'Title':job_title[:10],'Location':job_location[:10],'Company':company_name[:10],'Experience':experience_required[:10]})
df
```

```python
# Q4. Scrape data for first 100 sneakers you find when you visit flipkart.com and search for ‚Äúsneakers‚Äù inthe search field.

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
```

```python
#opening the flipkart webpage on automated Chrome browser.
driver.get("https://www.flipkart.com/")
```

```python
#entering product as 'sneakers' as required in question.
product=driver.find_element(By.CLASS_NAME,"Pke_EE")
product.send_keys("sneakers")
```

```python
#clicking on search button after entering product as 'sneakers'.
search=driver.find_element(By.CLASS_NAME,"_2iLD__")
search.click()
```

```python
#creating lists in which we are going to scrap data.
brand=[]
product_description=[]
price=[]
```

```python
#scraping the data of brand name from 3 pages
start=0
end=3
for page in range(start,end):
    brand_tags=driver.find_elements(By.XPATH,'//div[@class="syl9yP"]')
    for i in brand_tags:
        brand.append(i.text)
    next_button=driver.find_element(By.CLASS_NAME,"_9QVEpD") 
    next_button.click()
    time.sleep(3) 
```

```python
#scraping the data of product description from 4 pages
start=0
end=4
for page in range(start,end):
    description_tags=driver.find_elements(By.XPATH,'//a[@class="WKTcLC"]')
    for i in description_tags:
        product_description.append(i.text)
    next_button=driver.find_element(By.CLASS_NAME,"_9QVEpD") 
    next_button.click()
    time.sleep(4)
```

```python
#scraping the data of price from 3 pages
start=0
end=3
for page in range(start,end):
    price_tags=driver.find_elements(By.XPATH,'//div[@class="Nx9bqj"]')
    for i in price_tags:
        price.append(i.text.replace('‚Çπ',""))
    next_button=driver.find_element(By.CLASS_NAME,"_9QVEpD") 
    next_button.click()
    time.sleep(3)
```

```python
#creating the DataFrame from above scraped data.This dataframe is having data for 100 sneakers as required in question.
df=pd.DataFrame({'Brand':brand[:100],'Product_Description':product_description[:100],'Price':price[:100]})
df
```

```python
# Q6. Write a python program to scrape data for Top 1000 Quotes of All Time.

driver=webdriver.Chrome()
```

```python
#opening the azquotes webpage on automated Chrome browser.
driver.get("https://www.azquotes.com/")
```

```python
#Clicking on TopQuote.
Top_Quotes=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
Top_Quotes.click()
```

```python
#creating lists in which we are going to scrap data.
quote=[]
author=[] 
type_of_quotes=[]
```

```python
#scraping the data of 'Quotes' from 10 pages
start=0
end=10
for page in range(start,end):
    quote_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tags:
        quote.append(i.text)
    next_button=driver.find_element(By.CLASS_NAME,'next')
    next_button.click()
    time.sleep(10)
```

```python
#scraping the data of 'Authors' from 10 pages
start=0
end=10
for page in range(start,end):
    author_tags=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_tags:
        author.append(i.text)
    next_button=driver.find_element(By.CLASS_NAME,'next')
    next_button.click()
    time.sleep(10)
```

```python
#scraping the data of 'Type of quote' from 10 pages
start=0
end=10
for page in range(start,end):
    type_tags=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type_tags:
        type_of_quotes.append(i.text)
    next_button=driver.find_element(By.CLASS_NAME,'next')
    next_button.click()
    time.sleep(10)
```

```python
#creating the DataFrame from above scraped data.This dataframe is having data for 1000 Quotes as required in question.
df=pd.DataFrame({'Quote':quote[:1000],'Author':author[:1000],'Type_of_Quote':type_of_quotes[:1000]})
df
```
